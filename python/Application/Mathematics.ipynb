```python
# rvs 確率変数（）
norm_rvs = stats.norm.rvs(loc=50, scale=20, size=1000, random_state=0)
#期待値50, 標準偏差20, 個数1000
plt.hist(norm_rvs, bins=10, alpha=0.3, ec='blue')
plt.xlabel('number', fontsize=13)
plt.ylabel('count', fontsize=13)
plt.show()
```

```python
# pdf 確率密度関数(相対的な出やすさ)(連続型データ)
# 確率密度5の時の確率密度を計算

number = 5
y = stats.norm.pdf(x=number, loc=4, scale=0.8)
print("x=5 の確率密度：", y)

X = np.arange(start=1, stop=7.1, step=0.1)
norm_pdf = stats.norm.pdf(x=X, loc=4, scale=0.8)

plt.plot(X, norm_pdf)
plt.plot(number, y, 'bo')

plt.vlines(number, 0.0, y, lw=2, linestyles='dashed')
plt.hlines(y, 1.0, number, lw=2, linestyles='dashed')

plt.xlabel("number", fontsize=13)
plt.ylabel("density", fontsize=13)
plt.show()

#stats.norm.logpdf で対数グラフをかける。値だけなら pmf に log をつけても変わらない。
```

```python
# pmf 確率質量関数(離散型データ)(相対的な出やすさ)
# 2項分布　成功確率40%で5回試行して、そのうち成功する回数

p = 0.4 #成功確率
n = 5 #試行回数
k = np.arange(0, 6) #成功回数

binom_pmf = stats.binom.pmf(k, n, p)

plt.plot(k, binom_pmf, 'bo', ms=8)
plt.vlines(k, 0, binom_pmf, colors='b', lw=3, alpha=0.5)
plt.xticks(k)
plt.xlabel("number", fontsize=13)
plt.ylabel("pmf", fontsize=13)
plt.show()
```

```python
# cdf 累積分布関数　(x以下となる確率)

X = np.linspace(start=1, stop=7, num=100)
norm_cdf = stats.norm.cdf(x=X, loc=4, scale=0.8)
under_5 = stats.norm.cdf(x=5, loc=4, scale=0.8)
print("5以下になる確率：", under_5)

plt.plot(X, norm_cdf)
plt.plot(5, under_5, 'bo')

plt.vlines(5, 0.0, under_5, lw=2, linestyles='dashed')
plt.hlines(under_5, 1.0, 5, lw=2, linestyles='dashed')

plt.xlabel("number", fontsize=13)
plt.ylabel("cdf", fontsize=13)
plt.show()
```

```python
# sf 生存関数　(x以上となる確率)

X = np.linspace(start=1, stop=7, num=100)
norm_sf = stats.norm.sf(x=X, loc=4, scale=0.8)
upper_5 = stats.norm.sf(x=5, loc=4, scale=0.8)
print("5以上になる確率：", upper_5)

plt.plot(X, norm_sf)
plt.plot(5, upper_5, 'bo')

plt.vlines(5, 0.0, upper_5, lw=2, linestyles='dashed')
plt.hlines(upper_5, 1.0, 5, lw=2, linestyles='dashed')

plt.xlabel("number", fontsize=13)
plt.ylabel("sf", fontsize=13)
plt.show()
```

```python
# ppf パーセント点関数
#累積分布関数が「値-->割合」に対し、パーセント点関数は「割合-->値」

X = np.arange(start=1, stop=7, step=0.1)
norm_pdf = stats.norm.pdf(x=X, loc=4, scale=0.8)  # 確率密度関数を生成

q_75 = stats.norm.ppf(q=0.75, loc=4, scale=0.8)
print("累積分布関数７５％点の確率変数：", q_75)

v = stats.norm.pdf(x=q_75, loc=4, scale=0.8)
print("累積分布関数７５％点の確率密度：", v)

plt.plot(X, norm_pdf)
plt.plot(q_75, v, 'bo')

plt.vlines(q_75, 0.0, v, lw=2, linestyles='dashed')
plt.hlines(v, 1.0, q_75, lw=2, linestyles='dashed')

plt.xlabel("number", fontsize=13)
plt.ylabel("pdf", fontsize=13)
plt.show()
```

```python
# isf 逆生存関数 (ppfの逆)(sf の逆関数という意味)
X = np.arange(start=1, stop=7, step=0.1)
norm_pdf = stats.norm.pdf(x=X, loc=4, scale=0.8)  # 確率密度関数を生成

q_25 = stats.norm.isf(q=0.25, loc=4, scale=0.8)
print("逆生存関数２５％点の確率変数：", q_25)

v = stats.norm.pdf(x=q_25, loc=4, scale=0.8)
print("逆生存関数２５％点の確率密度：", v)

plt.plot(X, norm_pdf)
plt.plot(q_25, v, 'bo')

plt.vlines(q_25, 0.0, v, lw=2, linestyles='dashed')
plt.hlines(v, 1.0, q_25, lw=2, linestyles='dashed')

plt.xlabel("number", fontsize=13)
plt.ylabel("pdf", fontsize=13)
plt.show()
```

```python
# interval 区間推定
X = np.arange(start=1, stop=7, step=0.1)
norm_pdf = stats.norm.pdf(x=X, loc=4, scale=0.8)  # 確率密度関数を生成

lower, upper = stats.norm.interval(alpha=0.95, loc=4, scale=0.8)
print('信頼区間９５％の下限：', lower)
print('信頼区間９５％の上限：', upper)

v_lower = stats.norm.pdf(x=lower, loc=4, scale=0.8)
v_upper = stats.norm.pdf(x=upper, loc=4, scale=0.8)

plt.plot(X, norm_pdf)
plt.plot(lower, 0.0, 'k|')
plt.plot(upper, 0.0, 'k|')

plt.vlines(lower, 0.0, v_lower, lw=0.8)
plt.vlines(upper, 0.0, v_upper, lw=0.8)

plt.xlabel("number", fontsize=13)
plt.ylabel("pdf", fontsize=13)
plt.show()
```

```python
# ベルヌーイ分布(1回のベルヌーイ試行における確率分布)
# ベルヌーイ試行(2種類しかない事象のどちらかを発生させること)

x = np.array([0, 0, 1, 1, 0, 1, 0, 0]) #２種類の事象

p = len(x[x==1]) / len(x)
pmf_bernoulli = stats.bernoulli.pmf(x, p)

plt.vlines(x, 0, pmf_bernoulli, color='blue', lw=50)
plt.xticks([0, 1])
plt.xlim([0 - 0.5, 1 + 0.5])
plt.grid(True)

print("多い方：", pmf_bernoulli.max())
print("少ない方：", pmf_bernoulli.min())
```

```python
# ２項分布(独立したベルヌーイ試行を n 回繰り返した時の確率分布)
# 成功確率p=50%, 試行回数n=5回, 成功する回数k=2回
p = stats.binom.pmf(n=5, p=0.5, k=2)
print(f'{p:.4f}')
```

```python
# ポアソン分布(ごくまれに起こる事象の確率)(サンプルが多く、確率が非常に小さい)
# ある期間の平均発生回数(強度)5回の事象が2回起こる確率
stats.poisson.pmf(k=2, mu=5)
```

```python
# 数値積分
import numpy as np
from scipy import integrate
```

```python
# 普通の積分
y = lambda x: np.sin(x)**2
integ = integrate.quad(y, 0, np.pi)
print(integ)    # 0からπまでの積分、真値はπ/２
```

```python
# 無限積分
y = lambda x: np.exp(-x)
val, err = integrate.quad(y, 0, np.inf)
print(val)
print(err)
```

```python
# 積分変数以外に変数を含む
y = lambda x, a: 1/(x**2 + a**2)
val, err = integrate.quad(y, 0, 1, args=2)  # a=2 を設定している
print(val)
print(err)
```

```python
# 特異点を含む(コーシーの主値積分)  真値0
y = lambda x: 1
val, err = integrate.quad(y, -1, 1, weight='cauchy', wvar=0)  # cauchy にすると、1/(x-c) になり、cの値をwvarで与える。
print(val)

y2 = lambda x: 1/x
val2, err2 = integrate.quad(y2, -1, 1, points=0)  # 特異点の値、またはリストを渡す。
print(val2)
```

```python
# 重積分 
f = lambda y, x: (x**2)*y
val, err = integrate.dblquad(f, 0, 1, lambda x: 0, lambda x: 1)
print(val) # 1/6
print(err)

print()

f2 = lambda z, y, x: x + y + z
val, err = integrate.tplquad(f2, 0, 1,
                             lambda x: 0, lambda x: 1,
                             lambda x, y: 0, lambda x, y: 1)
print(val)
print(err)
```

```python
# 余因子(行列aからi行j列を除く)
def cofactor(a, i, j, minor=False):  # minor = True にすると小行列式を返す
    x = np.delete(a, i, axis=0)
    x = np.delete(x, j, axis=1)
    if minor == False:
        a_ij = (-1)**(i+j)*np.linalg.det(x)
    else:
        a_ij = np.linalg.det(x)
    return a_ij
# 余因子行列
for i in range(n):
    for j in range(m):
        a_ij[m, n] = cofactor(a, i, j)
```

```python
# 対角化
def eig_diag(x):
    eig = np.linalg.eig(x)
    e = np.diag(eig[0])     # 固有値の対角行列
    p = eig[1]              # 固有ベクトル
    return e, p
a = np.array([[3, 1], 
              [2, 4]])
e = eig_diag(a)
d = e[0]
p = e[1]
print(d)
pdp = np.dot(linalg.inv(p), np.dot(a, p))
print(pdp)   #  =a
```

```python
# 基本行列変形
def matrix_p(n, i=1, j=1):   # i行とj行の交換
    arr = np.eye(n)
    arr[[i-1, j-1]] = arr[[j-1, i-1]]
    return arr

def matrix_c(n, i=1, k=1):   # i行の定数倍
    arr = np.eye(n)
    arr[i-1, i-1] = k
    return arr

def matrix_e(n, i=1, j=1, k=1):   # i行にj行の定数倍を加える
    arr = np.eye(n)
    arr[i-1, j-1] = k
    return arr

p = matrix_p(4, 2, 3) # 4*4の行列 2行と3行の交換
print(p)
# x[[1, 2]] = x[[2, 1]]
print()

c = matrix_c(4, 3, 5)  # 4*4の行列 3行を5倍
print(c)
# x[2, :] = 5 * x[2, :]
print()

e = matrix_e(4, 3, 1, 2)  # 4*4の行列 3行目に1行目の2倍を加える。
print(e)
# x[2, :] = x[2, :] + 2 * x[0, :] 

inv_A = np.linalg.multi_dot([m7, m6, m5, m4, m3, m2, m1])
```
