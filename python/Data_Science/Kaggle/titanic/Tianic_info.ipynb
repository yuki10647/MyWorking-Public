```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib  inline
```

```python
train.groupby('Sex')[['Survived']].mean()
```

```python
#train.groupby(['Sex', 'Pclass'])['Survived'].aggregate('mean').unstack()
train.groupby(['Sex', 'Pclass'])[['Survived']].mean().unstack()
```

```python
age = pd.cut(train['Age'], [0, 18, 80])
train.pivot_table('Survived', ['Sex', age], 'Pclass')
```

```python
fare = pd.qcut(train['Fare'], 2)
train.pivot_table('Survived', ['Sex', age], [fare, 'Pclass'])
```

```python
train.pivot_table(index='Sex', columns='Pclass', aggfunc={'Survived':sum, 'Fare':'mean'})
```
